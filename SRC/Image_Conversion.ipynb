```python
from PIL import Image
import os

def convert_jpg_to_png(directory):
    # Create a new directory to save the converted images
    new_directory = '/content/drive/MyDrive/JPG CONV' + '/png'
    if not os.path.exists(new_directory):
        os.makedirs(new_directory)

    # Loop through all files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.jpg'):
            # Open the JPG image
            image = Image.open(os.path.join(directory, filename))

            # Convert and save the image as PNG
            png_filename = os.path.splitext(filename)[0] + '.png'
            image.save(os.path.join(new_directory, png_filename), 'PNG')

            print(f'Converted {filename} to {png_filename}')

    print('Conversion complete!')

# Specify the directory containing the JPG images
image_directory = '/content/drive/MyDrive/ocelot2023_v0.1.2/images/train/tissue'

# Call the conversion function
convert_jpg_to_png(image_directory)
```
